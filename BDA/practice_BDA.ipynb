linear regration

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = np.array(df['Manufacturing'])
Y = np.array(df['Wholesale'])
```

```python
coefficients = np.polyfit(X, Y, 1)

# Extract the slope and intercept from the coefficients
slope = coefficients[0]
intercept = coefficients[1]

# Print the coefficients
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
Yr = intercept + slope * X
```

```python
plt.title('Regration')
plt.scatter(X,Y)
plt.plot(X,Yr)
plt.xlabel('Manufacturing (thousands of people)')
plt.ylabel('Wholesale (thousands of people)')
plt.show()
```

```python
model = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=1/3,random_state=42)
model.fit(X_train.reshape(-1,1),Y_train)
```

```python
Ypred_train = model.predict(X_train.reshape(-1,1))
Ypred_test = model.predict(X_test.reshape(-1,1))
```

```python
error_train = mean_squared_error(Y_train,Ypred_train)
error_test = mean_squared_error(Y_test,Ypred_test)
```

```python
print(error_train)
print(error_test)
```

```python
plt.title('Regration')
plt.scatter(X,Y)
plt.plot(X,Yr)
plt.scatter(X_train,Ypred_train)
plt.legend(['Data', 'Regression Line', 'Predictions'])
plt.xlabel('Manufacturing (thousands of people)')
plt.ylabel('Wholesale (thousands of people)')
plt.show()
```
